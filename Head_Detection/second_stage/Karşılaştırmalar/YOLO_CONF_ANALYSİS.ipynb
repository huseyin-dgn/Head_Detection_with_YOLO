## Counting Odaklı CONF Analizi (CONF = 0.37, 0.39, 0.41, 0.43)

Bu bölümde YOLO tabanlı head detection modelinin farklı confidence (CONF) eşiklerinde
elde edilen sonuçlar **yalnızca sayım doğruluğu (counting accuracy)** perspektifiyle
değerlendirilmiştir. Amaç, görüntülerdeki kafa sayısını sistematik fazla veya eksik
saymadan mümkün olan en doğru şekilde tahmin etmektir.

### Kullanılan Veri
- **N_images:** 700
- **GT_total:** 18,605 (tüm test setindeki gerçek kafa sayısı)

---

### Özet Tablo

| CONF | MAE ↓ | RMSE ↓ | MAPE ↓ | Pred_total | Bias (Pred − GT) |
|------|------|------|------|------------|------------------|
| 0.37 | **1.656** | 4.447 | 0.07149 | 18,746 | **+141** |
| 0.39 | 1.669 | 4.361 | **0.07026** | 18,623 | **+18** |
| 0.41 | 1.693 | 4.201 | 0.07174 | 18,488 | **−117** |
| 0.43 | 1.731 | **4.119** | 0.07037 | 18,349 | **−256** |

---

### CONF = 0.37 Analizi
- En düşük **MAE (1.656)** değeri bu noktada elde edilmiştir.
- Ancak **Pred_total = 18,746**, yani model **+141 kafa fazla** saymaktadır.
- Bu durum, düşük conf değerinde **false positive** oranının arttığını ve özellikle
kalabalık sahnelerde duplicate detection riskinin yükseldiğini göstermektedir.

**Yorum:** Ortalama hata düşük olsa da, sistematik fazla sayım nedeniyle saf counting
uygulamaları için risklidir.

---

### CONF = 0.39 Analizi
- **MAE = 1.669**, CONF=0.37’ye çok yakın ve hâlâ düşük bir değerdedir.
- **Pred_total = 18,623**, yani yalnızca **+18 kafa** fazla sayım vardır.
- Görüntü başına ortalama bias ≈ **+0.026 kafa** olup pratikte ihmal edilebilir düzeydedir.
- RMSE ve MAPE dengeli bir seviyededir.

**Yorum:** Bu eşik, overcount ve undercount arasında **en dengeli noktayı** temsil eder.
Sistematik bias neredeyse sıfırdır.

---

### CONF = 0.41 Analizi
- MAE ve RMSE değerleri düşmeye devam etse de,
- **Pred_total = 18,488** ile **−117 kafa eksik** sayım başlamıştır.
- Bu, confidence arttıkça **false negative** etkisinin ağır bastığını göstermektedir.

**Yorum:** Sayım tarafında undercount eğilimi başladığı için dikkatli kullanılmalıdır.

---

### CONF = 0.43 Analizi
- En düşük **RMSE (4.119)** bu noktadadır; büyük hatalar bastırılmıştır.
- Ancak **Pred_total = 18,349**, yani **−256 kafa eksik** sayılmıştır.
- Bu durum, sistematik undercount anlamına gelir.

**Yorum:** Outlier’ları azaltır ancak toplam sayım doğruluğu açısından uygun değildir.

---

### Genel Değerlendirme ve Sonuç
- **CONF = 0.37**: En düşük MAE, ancak belirgin overcount.
- **CONF = 0.43**: Düşük RMSE, ancak belirgin undercount.
- **CONF = 0.41**: Undercount eğilimi başlamaktadır.
- **CONF = 0.39**: MAE, RMSE ve MAPE dengeli; **toplam sayım bias’ı en düşük**.

**Sonuç:**  
Counting (kaç kafa var?) problemi için **CONF = 0.39**, sistematik fazla veya eksik
sayımı minimize eden ve en dengeli sayım performansını sunan optimal eşik olarak
seçilmiştir.
