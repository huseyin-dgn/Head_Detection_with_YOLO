```python
import os, shutil, json, math
from pathlib import Path
import pandas as pd
import numpy as np
from ultralytics import YOLO
import torch
```

```python
DATASET_DIR = Path(r"C:\Users\hdgn5\OneDrive\Masaüstü\Head Detection\scut_head_yolo_rebalanced")
DATA_YAML   = DATASET_DIR / "dataset.yaml"

FINAL_DIR   = Path(r"C:\Users\hdgn5\OneDrive\Masaüstü\Head Detection\Final")
RUNS_DIR    = FINAL_DIR / "runs"
REPORTS_DIR = FINAL_DIR / "reports"
SCRIPTS_DIR = FINAL_DIR / "scripts"
BACKUP_DIR  = FINAL_DIR / "weights_backup"

RUNS_DIR.mkdir(parents=True, exist_ok=True)
REPORTS_DIR.mkdir(parents=True, exist_ok=True)
SCRIPTS_DIR.mkdir(parents=True, exist_ok=True)
BACKUP_DIR.mkdir(parents=True, exist_ok=True)

print("DATA_YAML:", DATA_YAML)
print("FINAL_DIR:", FINAL_DIR)
```

```python
ORIGINAL_BEST_PT = Path(r"C:\Users\hdgn5\OneDrive\Masaüstü\Head Detection\Final\weights\best.pt")
assert ORIGINAL_BEST_PT.exists(), f"best.pt yok: {ORIGINAL_BEST_PT}"

backup_path = BACKUP_DIR / "best_original.pt"
shutil.copy2(ORIGINAL_BEST_PT, backup_path)
print("Backup OK ->", backup_path)

model = YOLO(str(backup_path))
print("Model loaded.")
```

```python
print("CUDA available:", torch.cuda.is_available())
if torch.cuda.is_available():
    print("GPU:", torch.cuda.get_device_name(0))
```

```python
train_run_name = "B_v8n_finetune_img960_ep60_lr002_bs2"

model = YOLO(str(backup_path))

train_results = model.train(
    data=str(DATA_YAML),
    imgsz=960,
    epochs=30,
    batch=2,
    device=0,
    workers=2,
    project=str(RUNS_DIR),
    name=train_run_name,
    exist_ok=False,
    patience=15,
    amp=True,
    lr0=0.002,
    lrf=0.01,
    weight_decay=0.0005,
    close_mosaic=5,
    degrees=0.0,
    perspective=0.0
)
print("TRAIN DONE ->", RUNS_DIR / train_run_name)
```
