## 2. Model Eğitimi

```python
from ultralytics import YOLO

best_pt = r"C:\Users\hdgn5\OneDrive\Masaüstü\Head Detection\head_yolo_project\models\baseline_y8s_768_adamw_lr25e-4\best.pt"

data_yaml = r"C:\Users\hdgn5\OneDrive\Masaüstü\Head Detection\scut_head_yolo_rebalanced\dataset.yaml"

model = YOLO(best_pt)

results = model.train(
    data=data_yaml,
    imgsz=832,
    epochs=20,
    batch=8,          
    device=0,
    workers=4,        
    seed=42,

    optimizer="AdamW",
    lr0=0.001,
    lrf=0.01,
    weight_decay=0.0005,

    cos_lr=True,
    close_mosaic=0,   # fine-tune'da mosaic kapalı
    patience=10,

    pretrained=True,
    amp=True,

    project=r"C:\Users\hdgn5\OneDrive\Masaüstü\Head Detection\head_yolo_project\outputs\runs",
    name="finetune_y8s_832_lr1e-3_nomosaic",
    exist_ok=True,
    save=True
)
print("Done:", results.save_dir)
```
