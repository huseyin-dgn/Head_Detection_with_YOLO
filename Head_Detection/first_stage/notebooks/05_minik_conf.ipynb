```python
from pathlib import Path
import os, sys

# Eğer eski path çakışması varsa temizle
sys.path = [p for p in sys.path if p and p != r"C:\HysoExpect"]

PROJECT_ROOT = Path.cwd().parent
sys.path.insert(0, str(PROJECT_ROOT))
os.chdir(PROJECT_ROOT)

best_pt = Path(r"C:\Users\hdgn5\OneDrive\Masaüstü\Head Detection\outputs\runs\finetune_y8s_832_lr1e-3_nomosaic\weights\best.pt")
dataset_root = Path(r"C:\Users\hdgn5\OneDrive\Masaüstü\Head Detection\scut_head_yolo_rebalanced")
test_images = dataset_root / "images" / "test"
test_labels = dataset_root / "labels" / "test"

base_out = Path(r"C:\Users\hdgn5\OneDrive\Masaüstü\Head Detection")

from src.predict import predict_folder
from src.eval_count import eval_count

def run_conf(conf: float, tag: str, iou: float = 0.6, imgsz: int = 832, device: str = "cuda:0"):
    out_root = base_out / f"outputs_{tag}"
    pred_out = out_root / "predictions"
    rep_out  = out_root / "reports"

    # 1) Tahminleri üret
    predict_folder(
        weights=str(best_pt),
        images_dir=str(test_images),
        out_dir=str(pred_out),
        conf=conf,
        iou=iou,
        imgsz=imgsz,
        device=device,
    )

    # 2) Count eval + summary
    df, summary = eval_count(
        weights=str(best_pt),
        images_dir=str(test_images),
        labels_dir=str(test_labels),
        report_dir=str(rep_out),
        conf=conf,
        iou=iou,
        imgsz=imgsz,
        device=device,
    )

    print(f"\n=== CONF {conf:.2f} | IOU {iou:.2f} | imgsz {imgsz} | device {device} (outputs_{tag}) ===")
    print(summary)
    return df, summary


# İSTEDİĞİN CONF DEĞERLERİ
conf_list = [0.37, 0.39, 0.41, 0.43]

results = {}
for conf in conf_list:
    # float sapmasına dayanıklı tag
    tag = f"conf{int(round(conf*100)):02d}_iou60"  # conf37_iou60, conf39_iou60, ...
    df, summary = run_conf(conf, tag, iou=0.6, imgsz=832, device="cuda:0")
    results[conf] = {"df": df, "summary": summary}

print("\n\n========== ALL SUMMARIES ==========")
for conf in conf_list:
    print(f"\n--- CONF {conf:.2f} ---")
    print(results[conf]["summary"])
```
